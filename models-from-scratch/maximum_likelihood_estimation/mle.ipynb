# Estimativa por Máxima Verossimilhança

A Estimativa por Máxima Verossimilhança é um método estatístico utilizado para estimar parâmetros de uma distribuição de variáveis onde apenas os dados de alguma amostra da população está disponível, supondo uma distribuição de probabilidade inicial. Como exemplo, pode-se tomar o desafio de calcular a largura de folhas de determinada planta - é impossível coletar todas as folhas do mundo, temos que nos contentar com a medição de apenas algumas amostras.

## 1. Suposição de uma distribuição inicial

Essa é a parte em que o conhecimento do domínio entra na jogada. Existem diversos tipos de distribuições: normal, uniforme, exponencial, gamma... uma infinidade. Qual a distribuição de probabilidades que melhor se adequa aos seus dados coletados? Neste caso, escolheremos a distribuição normal, expressa pela equação abaixo.

<center>
$ P(x | \mu, \sigma) = \dfrac{1}{\sigma \sqrt{2 \pi}} e^{\dfrac{1}{2}\left(\dfrac{x-\mu}{\sigma} \right)^{2}}$
</center>

Numa distribuição normal temos dois parâmetros de ajuste: a média ($\mu$) e o desvio padrão ($\sigma$). Nela, espera-se que a maioria das medidas estejam perto e relativemente simétricas em relação à média.

## 2. Função de verossimilhança 

O que queremos responder com o método da máxima verossimilhança é: dado um conjunto de parâmetros $\theta$ de uma distribuição de probabilidades (onde numa normal é o conjunto $\{ \mu, \sigma \}$), qual a probabilidade de um conjunto de observações pertencerem àquela distribuição?

Definimos essa probabilidade a partir da função de verossimilhança, com estimador definido abaixo, onde $f$ é a função de probabilidade assumida inicialmente:

<center>
    $ \hat{\mathcal{L}}(\theta;x) = \dfrac{1}{n}\sum_{i=1}^{n} \ln{f(x_i|\theta)} $
</center>
